Importing the data from saved csv files

```python
def get_stored_data(end_date, num_days, path=''):
    # Get the dates we want to learn about
    date_list = pd.date_range(end=end_date, periods=num_days, freq='B')

    # Get the corresponding files for the dates
    dates = []
    data = []

    # Go through the dates and add files if they exist
    for d in date_list:
        if os.path.exists(path + 'csv_files/company_stats_'+ str(d.date()) + '.csv'):
            try:
                data.append(pd.read_csv(path + 'csv_files/company_stats_'+ str(d.date()) + '.csv', encoding='cp1252'))
                dates.append(d)
            except:
                print('Data import failed for ' + 'csv_files/company_stats_'+ str(d.date()) + '.csv')
    return data, dates
```

```python
data, dates = get_stored_data('05-07-2020', 120)
```

```python
def get_metric_from_list_of_time_series_data(data, value_retriever, statistic_function):
    return [statistic_function(value_retriever(x)) for x in data]

def get_metric_over_time(data, column, trim_mean_thresh=0.1):
    return get_metric_from_list_of_time_series_data(data, lambda x: list(x[column]), lambda x: trim_mean([i for i in x if str(i) != 'nan'], trim_mean_thresh))

def plot_metric(data, dates, column, trim_mean_thresh=0.1):
    price_plot(dates, get_metric_over_time(data, column, trim_mean_thresh=trim_mean_thresh), title=column)
```

```python
plot_metric(data, dates, 'Diluted EPS')
plot_metric(data, dates, 'Price')
plot_metric(data, dates, 'Profit Margin')
plot_metric(data, dates, 'EPS Beat Ratio')
plot_metric(data, dates, 'Total Debt/Equity')
plot_metric(data, dates, 'Current Ratio')
plot_metric(data, dates, 'EBITDA')
```
